# Resultados de los expermientos

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
````

````python
#Dataframe preguntas y respuestas que se esperan de los modelos
df_preguntas_respuestas = pd.read_csv("preguntas_respuestas.csv")
df_preguntas_respuestas
````

````python
df_resultados_experimentos = pd.read_csv("resultados_experimentos_two.csv")
df_resultados_experimentos
````

````python
# Grafiquemos la distribucion del tiempo

plt.figure(figsize=(10, 6))
plt.hist(df_resultados_experimentos['Tiempo_Respuesta'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribución del Tiempo de Respuesta de los Modelos')
plt.xlabel('Tiempo de Respuesta (segundos)')
plt.ylabel('Frecuencia')
plt.grid(axis='y', alpha=0.75)
plt.show()
````

````python
# === Lista de etiquetas para las preguntas ===
preguntas = [
    "Pregunta 1", "Pregunta 2", "Pregunta 3", "Pregunta 4", "Pregunta 5",
    "Pregunta 6", "Pregunta 7", "Pregunta 8", "Pregunta 9", "Pregunta 10",
    "Pregunta 11", "Pregunta 12", "Pregunta 13", "Pregunta 14", "Pregunta 15"
]

# === Gráficas comparativas: todos los modelos para un mismo prompt ===
prompts = df_resultados_experimentos["Prompt"].unique()

for prompt in prompts:
    plt.figure(figsize=(20, 9))
    df_prompt = df_resultados_experimentos[df_resultados_experimentos["Prompt"] == prompt]

    for modelo in df_prompt["Modelo"].unique():
        df_modelo = df_prompt[df_prompt["Modelo"] == modelo]
        plt.plot(
            range(1, len(df_modelo) + 1),
            df_modelo["Tiempo_Respuesta"],
            marker='o',
            linestyle='-',
            label=modelo
        )

    plt.xticks(range(1, len(df_prompt["Pregunta"].unique()) + 1),
               preguntas[:len(df_prompt["Pregunta"].unique())],
               rotation=45)
    plt.xlabel("Preguntas")
    plt.ylabel("Tiempo de respuesta (s)")
    plt.title(f"Comparativa de modelos - {prompt}")
    plt.legend(loc="upper left")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
````

````python
# === 3. Gráfica de distribución de tiempos por Prompt (todos los modelos) ===
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_resultados_experimentos, x="Prompt", y="Tiempo_Respuesta")
plt.title("Distribución de tiempos de respuesta por Prompt (todos los modelos)")
plt.xlabel("Prompt")
plt.ylabel("Tiempo de respuesta (s)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

````python
# === 4. Gráfica de distribución de tiempos por Prompt y Modelo ===
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_resultados_experimentos, x="Prompt", y="Tiempo_Respuesta", hue="Modelo")
plt.title("Distribución de tiempos de respuesta por Prompt y Modelo")
plt.xlabel("Prompt")
plt.ylabel("Tiempo de respuesta (s)")
plt.legend(title="Modelo")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

````python
# === 3. Crear una columna con los nombres cortos ===
# Asumimos que el orden en el CSV es el mismo que la lista de preguntas
# Si no lo es, habría que mapearlas explícitamente
unique_preguntas = df_resultados_experimentos["Pregunta"].unique()
mapa_preguntas = {p: preguntas[i] for i, p in enumerate(unique_preguntas)}
df_resultados_experimentos["Pregunta_Corta"] = df_resultados_experimentos["Pregunta"].map(mapa_preguntas)

# === 4. Configurar estilo de las gráficas ===
sns.set(style="whitegrid", palette="pastel")

# === 5. Gráfica de distribución de tiempos por pregunta (independiente de modelo/prompt) ===
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_resultados_experimentos, x="Pregunta_Corta", y="Tiempo_Respuesta")

plt.title("Distribución de tiempos de respuesta por pregunta (todos los modelos y prompts)")
plt.xlabel("Pregunta")
plt.ylabel("Tiempo de respuesta (s)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````
